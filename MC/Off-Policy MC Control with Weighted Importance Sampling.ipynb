```python
%matplotlib inline

import gym
import matplotlib
import numpy as np
import sys

from collections import defaultdict
if "../" not in sys.path:
  sys.path.append("../") 
from lib.envs.blackjack import BlackjackEnv
from lib import plotting

matplotlib.style.use('ggplot')
```

```python
env = BlackjackEnv()
```

```python
def create_random_policy(nA):
    """
    Creates a random policy function.
    
    Args:
        nA: Number of actions in the environment.
    
    Returns:
        A function that takes an observation as input and returns a vector
        of action probabilities
    """
    A = np.ones(nA, dtype=float) / nA
    def policy_fn(observation):
        return A
    return policy_fn
```

```python
def create_greedy_policy(Q):
    """
    Creates a greedy policy based on Q values.
    
    Args:
        Q: A dictionary that maps from state -> action values
        
    Returns:
        A function that takes an observation as input and returns a vector
        of action probabilities.
    """
    
    def policy_fn(state):
        # get a list of all the action values possible from the observation
        action_values = Q[state]
        # Create an array that stores the list of actoins
        A = np.zeros_like(action_values, dtype=float)
        # choose the best action from the list of action_values possible from the given observation
        best_action = np.argmax(action_values)
        A[best_action] = 1.0
        return A

    return policy_fn
```

```python
def mc_control_importance_sampling(env, num_episodes, behavior_policy, discount_factor=1.0):
    """
    Monte Carlo Control Off-Policy Control using Weighted Importance Sampling.
    Finds an optimal greedy policy.
    
    Args:
        env: OpenAI gym environment.
        num_episodes: Number of episodes to sample.
        behavior_policy: The behavior to follow while generating episodes.
            A function that given an observation returns a vector of probabilities for each action.
        discount_factor: Gamma discount factor.
    
    Returns:
        A tuple (Q, policy).
        Q is a dictionary mapping state -> action values.
        target_policy is a function that takes an observation as an argument and returns
        action probabilities. This is the optimal greedy policy.
    """
    
    # The final action-value function.
    # A dictionary that maps state -> action values
    Q = defaultdict(lambda: np.zeros(env.action_space.n))
    # The cumulative denominator of the weighted importance sampling formula
    # (across all episodes)
    C = defaultdict(lambda: np.zeros(env.action_space.n))
    
    # Our greedily policy we want to learn
    target_policy = create_greedy_policy(Q)

    
    # iterate through all the episodes
    for i_episode in range(num_episodes):
        # show a visual update for which episode we are on
        if i_episode % 1000 == 0:
            print("\rEpisode {}/{}.".format(i_episode, num_episodes), end="")
            sys.stdout.flush()
        # Initialize the starting state and episode list
        state = env.reset()
        episode = []

        # iterate through all the events in the current episode
        while True:
            # get information from the environment with respect to our behavior policy
            action_probs = behavior_policy(state)
            # choose a random action from probability list
            action = np.random.choice(np.arange(len(action_probs)), p=action_probs)
            next_state, reward, done, _ = env.step(action)
            episode.append((state, action, reward))
            # stopping condition
            if done: break
            state = next_state

        # Process the episode information into our Q and update target policy accordingly 
        # NOTE that we have collected all the episodic information with respect to the behavior policy for right now but we 
        # can use the importance sample ratio to transform the expected return of the behavior policy into the expected return of the target policy
        # initialize the sum of discounted returns according to the behavior policy
        G = 0
        importance_sample_ratio = 1.0

        # collect and iterate through all the events in the episode in reverse order for this epsisode
        for t in range(len(episode))[::-1]:
            state, action, reward = episode[t]
            # Update the total reward since step t
            G = (discount_factor * G) + reward
            # Update weighted importance sampling formula denominator
            C[state][action] += importance_sample_ratio
            # Update the action-value function using the incremental update formula (5.7)
            # This also improves our target policy which holds a reference to Q
            Q[state][action] += (importance_sample_ratio / (C[state][action])) * (G - Q[state][action])
            # If the action taken by the behavior policy is not the action 
            # taken by the target policy the probability will be 0 and we can break
            if action != np.argmax(target_policy(state)): break
            importance_sample_ratio = importance_sample_ratio * 1./behavior_policy(state)[action]

    
    return Q, target_policy
```

```python
random_policy = create_random_policy(env.action_space.n)
Q, policy = mc_control_importance_sampling(env, num_episodes=500000, behavior_policy=random_policy)
```

```python
# For plotting: Create value function from action-value function
# by picking the best action at each state
V = defaultdict(float)
for state, action_values in Q.items():
    action_value = np.max(action_values)
    V[state] = action_value
plotting.plot_value_function(V, title="Optimal Value Function")
```
